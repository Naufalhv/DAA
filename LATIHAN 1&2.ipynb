```python
import networkx as nx
import matplotlib as plt
```

```python
#LATIHAN 1

vertices = range(0, 11)
edges = [
    (1, 2), (1, 7),
    (2, 3), (2, 6), (2, 1),
    (3, 6), (3, 4), (3, 2),
    (4, 3),
    (5, 6),
    (6, 7), (6, 2), (6, 3),
    (7, 8), (7, 9), (7, 10), (7, 6), (7, 10),
    (8, 9), (8, 7),
    (9, 8), (9, 7), 
    (10, 7)
]
node_f = [1, 4, 5, 7, 10]
node_nf = [2, 3, 6, 8, 9]

G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
pos = nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G, pos, nodelist=node_f, label=True, node_color='r', node_size=1300)
nx.draw_networkx_nodes(G, pos, nodelist=node_nf, label=True, node_color='g', node_size=1300)
nx.draw_networkx_edges(G, pos, edges, width=3, alpha=0.5, edge_color='b')
labels = {}
labels[1] = r'1 F'
labels[2] = r'2 NF'
labels[3] = r'3 NF'
labels[4] = r'4 F'
labels[5] = r'5 F'
labels[6] = r'6 NF'
labels[7] = r'7 F'
labels[8] = r'8 NF'
labels[9] = r'9 NF'
labels[10] = r'10 F'

def calculate_dos(node):
    dos = 0
    for neighbor in G.neighbors(node):
        if labels[neighbor] != labels[node]:
            dos += 1
    return dos

dos_values = {node: calculate_dos(node) for node in G.nodes()}
labels = {node: f'{node} ({dos_values[node]})' for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=16)
```
